```python
import os

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.decomposition import PCA

# Set Seaborn style for all plots
sns.set(style="whitegrid", context="talk")
```

```python
root_dir = os.path.dirname(os.getcwd())
data_path = os.path.join(root_dir, 'data/processed_data.pkl')
df = pd.read_pickle(data_path)
df
```

```python
def plot_tpm_distribution(df, tpm_column_name='tpm', species=None):
    if species:
        df = df[df['species_name'].isin(species)]
    plt.figure(figsize=(12, 8))
    sns.boxplot(x='species_name', y=tpm_column_name, data=df, palette='Spectral')
    plt.title('TPM Distribution Across Species', fontsize=20, fontweight='bold')
    plt.xlabel('Species Name', fontsize=16)
    plt.ylabel('TPM', fontsize=16)
    plt.xticks(fontsize=14, rotation=45)
    plt.yticks(fontsize=14)
    plt.show()
```

```python
plot_tpm_distribution(df)
```

```python
# Plot 2: Stress Condition Influence on TPM
def plot_stress_condition_influence(df, tpm_column_name='tpm', stress_conditions=None):
    if stress_conditions:
        df = df[df['stress_condition_name'].isin(stress_conditions)]
    plt.figure(figsize=(14, 8))
    sns.boxplot(x='stress_condition_name', y=tpm_column_name, data=df, palette='Paired')
    plt.title('Stress Condition Influence on TPM', fontsize=20, fontweight='bold')
    plt.xlabel('Stress Condition Name', fontsize=16)
    plt.ylabel('TPM', fontsize=16)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.show()
```

```python
plot_stress_condition_influence(df)
```

```python
# Plot 4: Species-Specific Stress Conditions
def plot_species_specific_stress_conditions(df, tpm_column_name='tpm'):
    plt.figure(figsize=(16, 10))
    sns.violinplot(x='stress_condition_name', y=tpm_column_name, hue='species_name', data=df, 
                   palette=palette_dict, split=True, inner="quartile", scale='width', bw=0.2, dodge=True)
    plt.title('Species-Specific Stress Conditions Influence on TPM', fontsize=20, fontweight='bold')
    plt.xlabel('Stress Condition Name', fontsize=16)
    plt.ylabel('TPM', fontsize=16)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.legend(title='Species Name', fontsize=12, title_fontsize=14, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., ncol=1)
    plt.tight_layout()
    plt.show()
```

```python
# Custom color palette for 29 species
species_list = df['species_name'].unique()
palette = sns.color_palette("Spectral", len(species_list))
palette_dict = dict(zip(species_list, palette))

stress_conditions = df.stress_condition_name.unique()
num_stress_conditions = len(stress_conditions)
```

```python
plot_species_specific_stress_conditions(df[df['stress_condition_name'].isin(stress_conditions[:num_stress_conditions//3])])
```

```python
plot_species_specific_stress_conditions(df[df['stress_condition_name'].isin(stress_conditions[num_stress_conditions//3:2 * (num_stress_conditions//3)])])
```

```python
plot_species_specific_stress_conditions(df[df['stress_condition_name'].isin(stress_conditions[2 * num_stress_conditions//3:])])
```

```python
# Function to normalize TPM within each stress condition group
def normalize_group(group):
    mean = group['tpm'].mean()
    std = group['tpm'].std()
    group[f'tpm_normalized'] = (group['tpm'] - mean) / std
    return group

# Apply the normalization
df = df.groupby('stress_condition_name').apply(normalize_group).reset_index(drop=True)
df = df.rename(columns={"tpm_normalized": "tpm_stress_normalized"})
df = df.groupby('species_name').apply(normalize_group).reset_index(drop=True)
df = df.rename(columns={"tpm_normalized": "tpm_species_normalized"})
```
